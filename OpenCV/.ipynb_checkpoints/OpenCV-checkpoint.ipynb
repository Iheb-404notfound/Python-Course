<h3>addWeighted

```python
black_donkey = cv2.imread("black_donkey.png",1)
lines = cv2.imread("lines.png",1)

white_and_black_zebra = cv2.addWeighted(black_donkey,1,lines,1,0)

cv2.imshow("Equus",white_and_black_zebra)
cv2.waitKey(0)
```

<img src="donkey.png"/><img src="lines.png"/><img src="inverted_zebra.jpg"/>

<h3>subtract

```python
donkey = cv2.imread("donkey.png",1)
lines = cv2.imread("lines.png",1)

black_and_white_zebra = cv2.subtract(donkey,lines)

cv2.imshow("Equus",black_and_white_zebra)
cv2.waitKey(0)
```

<img src="black_donkey.png"/><img src="lines.png"/><img src="zebra.png"/>

<h3>bitwise_and

```python
donkey = cv2.imread("donkey.png",1)
lines = cv2.imread("lines.png",1)

bitwise_and_zebra = cv2.bitwise_and(donkey,lines,mask=None)

cv2.imwrite("bitwise_and.png",bitwise_and_zebra)
cv2.imshow("Equus",bitwise_and_zebra)
cv2.waitKey(0)
```

<img src="bitwise_and.png"/>

<h3>resize

```python
foal = cv2.resize(black_and_white_zebra,(300,200))

cv2.imwrite("foal.png",foal)
cv2.imshow("Baby Zebra",foal)
cv2.waitKey(0)
```

<img src='foal.png' width="300">

<h3>shape

```python
foal.shape
```

<h3>warpAffine

```python
height = foal.shape[0]
width = foal.shape[1]
rotation_degrees = 60
centerx = width/2
centery = height/2

rotational_transformation_matrix = cv2.getRotationMatrix2D((centerx,centery),rotation_degrees,1)

unicorn = cv2.warpAffine(foal, rotational_transformation_matrix, (width,height))

cv2.imwrite("unicorn.png",unicorn)
cv2.imshow("The foal is flying",unicorn)
cv2.waitKey(0)
```

<img src='unicorn.png' width="300">

<h3>Canny

```python
detailed_pythagoras = cv2.Canny(statue_colored,100,200)

cv2.imwrite("detailed_pythagoras.png",detailed_pythagoras)
cv2.imshow("time²+patience²=future²",detailed_pythagoras)
cv2.waitKey(0)
```

<img src='detailed_pythagoras.png' width="300">

<h3>GaussianBlur

```python
fading_pythagoras = cv2.GaussianBlur(statue_colored,(9,9),0)

cv2.imwrite("gaussian_pythagoras.png",fading_pythagoras)
cv2.imshow("Pythagoras invented and believed in Pythagoreanism",fading_pythagoras)
cv2.waitKey(0)
```

<img src='gaussian_pythagoras.png' width="300">

<h3>medianBlur

```python
fading_pythagoras = cv2.medianBlur(statue_colored,11)

cv2.imwrite("median_pythagoras.png",fading_pythagoras)
cv2.imshow("Pythagoras believed that he had already lived four lives",fading_pythagoras)
cv2.waitKey(0)
```

<img src='median_pythagoras.png' width="300">

<h3>bilateralFilter

```python
fading_pythagoras = cv2.bilateralFilter(statue_colored, 20, 200,200)

cv2.imwrite("bilateral_pythagoras.png",fading_pythagoras)
cv2.imshow("Pythios agora is what Pythagoras is ahora",fading_pythagoras)
cv2.waitKey(0)
```
